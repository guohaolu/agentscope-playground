## 导入

```python
from agentscope.message import (
    Msg,
    Base64Source,
    TextBlock,
    ThinkingBlock,
    ImageBlock,
    AudioBlock,
    VideoBlock,
    ToolUseBlock,
    ToolResultBlock,
)
import json
```

## 文本消息

```python
msg = Msg(name="Jarvis", role="assistant", content="Hi! How can I help you?")
print(f"发送者 name: {msg.name}")
print(f"角色 role: {msg.role}")
print(f"内容 content: {msg.content}")
```

## 多模态消息（ContentBlock）

使用 TextBlock、ImageBlock、AudioBlock、VideoBlock，source 可为 URL 或 Base64Source。

```python
msg_multi = Msg(
    name="Jarvis",
    role="assistant",
    content=[
        TextBlock(type="text", text="这是一条带文本的多模态消息。"),
        # ImageBlock/AudioBlock/VideoBlock 用法见官方 doc
    ],
)
print(msg_multi.get_text_content())
```

## 思考块与工具消息

```python
msg_thinking = Msg(
    name="Jarvis",
    role="assistant",
    content=[
        ThinkingBlock(type="thinking", thinking="推理过程示例"),
        TextBlock(type="text", text="最终回复示例"),
    ],
)

msg_tool_call = Msg(
    name="Jarvis",
    role="assistant",
    content=[
        ToolUseBlock(type="tool_use", id="343", name="get_weather", input={"location": "Beijing"})
    ],
)
msg_tool_res = Msg(
    name="system",
    role="system",
    content=[
        ToolResultBlock(type="tool_result", id="343", name="get_weather", output="北京晴 25°C")
    ],
)
print("工具调用与结果消息已创建")
```

## 序列化与反序列化

```python
serialized = msg.to_dict()
print(json.dumps(serialized, indent=2, ensure_ascii=False))
new_msg = Msg.from_dict(serialized)
print(f"反序列化后 name: {new_msg.name}")
```
