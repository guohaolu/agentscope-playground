## 学习要点

- Memory 的 add / get_memory 等接口。
- 与 Formatter 结合构造 prompt 时的消息顺序。

```python
from agentscope.memory import InMemoryMemory
from agentscope.message import Msg
import asyncio

async def demo():
    mem = InMemoryMemory()
    await mem.add(Msg("user", "你好", "user"))
    await mem.add(Msg("assistant", "你好！有什么可以帮您？", "assistant"))
    history = await mem.get_memory()
    print(f"当前记忆条数: {len(history)}")

# 参考 https://doc.agentscope.io/tutorial/task_memory.html 深入
asyncio.run(demo())
```
