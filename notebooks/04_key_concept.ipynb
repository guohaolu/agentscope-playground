# 1.4 关键概念梳理

- **对应文档**: [quickstart_key_concept](https://doc.agentscope.io/tutorial/quickstart_key_concept.html)
- **本讲目标**: 理解 State、Message、Tool、Agent、Formatter、Long-Term Memory 在工程中的角色。
- **前置知识**: 01～03。

## 概念速览

| 概念 | 要点 |
|------|------|
| **State** | 有状态对象通过 `state_dict`/`load_state_dict` 做快照与恢复；Agent、Memory、Toolkit 等均支持。 |
| **Message** | 统一数据结构：name, role, content, metadata；用于 Agent 间通信、UI、记忆、LLM API 媒介。 |
| **Tool** | 任意可调用对象（函数/方法/偏函数等），支持 async/sync、streaming/non-streaming。 |
| **Agent** | AgentBase: `reply`/`observe`/`print`；ReActAgentBase 增加 `_reasoning`、`_acting`。 |
| **Formatter** | 将 Message 转为 LLM API 格式；多身份场景用 multi-agent formatter。 |
| **Long-Term Memory** | 无严格长短记忆区分；支持 agent_control/static_control 等模式。 |

```python
from agentscope.message import Msg, ThinkingBlock, TextBlock
# 示例：检查有状态对象是否支持 state_dict
from agentscope.memory import InMemoryMemory

mem = InMemoryMemory()

msg_thinking = Msg(
    name="Jarvis",
    role="assistant",
    content=[
        ThinkingBlock(type="thinking", thinking="推理过程示例"),
        TextBlock(type="text", text="最终回复示例"),
    ],
)

await mem.add(memories=msg_thinking)

state = mem.state_dict()
print("InMemoryMemory 支持 state_dict:", isinstance(state, dict))
```
